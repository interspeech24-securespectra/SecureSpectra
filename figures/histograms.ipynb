```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from swarm_visualizer.utility.general_utils import save_fig, set_plot_properties
```

```python
import os
os.getcwd()
```

```python
cv16_original = np.load("../commonvoice_experiments/cv_16_original.npy")
cv16_openvoice = np.load("../commonvoice_experiments/cv_16_open_voice.npy")
cv16_tts = np.load("../commonvoice_experiments/cv_16_coquiTTS.npy")
cv16_whisper = np.load("../commonvoice_experiments/cv_16_whisperspeech.npy")
```

```python
set_plot_properties()
fig, (ax, ax2) = plt.subplots(2, 1, sharex=True, figsize=(12, 4), gridspec_kw={'height_ratios': [1, 4], 'hspace': 0.04})

# plt.figure(figsize=(12, 4))  # Optional: Adjust the figure size
set_plot_properties()
data = {
    'Frequency Bins (kHz)': np.tile(np.round(np.arange(0, 11)*0.6+5.4,2), 4),  # Repeat the category labels for each group
    'Normalized Mean Energy': np.concatenate([cv16_original, cv16_tts, cv16_openvoice, cv16_whisper]),
    'Group': ['Original'] * 11 + ['TTS'] * 11 + ['OpenVoice'] * 11 + ['WhisperSpeech'] * 11
}

df = pd.DataFrame(data)

# Create the plot
sns.barplot(data=df, x='Frequency Bins (kHz)', y='Normalized Mean Energy', hue='Group', dodge=True, ax=ax)
sns.barplot(data=df, x='Frequency Bins (kHz)', y='Normalized Mean Energy', hue='Group', dodge=True, ax=ax2)

# Set the y-axis limits to create the break effect
ax.set_ylim(0.975, 1)  # Top plot for higher values
ax2.set_ylim(0, 0.0045)   # Bottom plot for lower values

# Hide the spines between ax and ax2
ax.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)

# Add diagonal lines to indicate the break
d = .0105  # how big to make the diagonal lines in axes coordinates
kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
ax.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal
#ax.legend(False)
ax2.legend([],[], frameon=False)
ax.legend([],[], frameon=False)
# Display the plot



ax.set_title("Mozilla CommonVoice Dataset")
ax2.legend(title='DF Models', loc=(1.025,0))
ax.set_ylabel("")
plt.savefig("mozilla_hist.png", dpi=600)
plt.show()
```
